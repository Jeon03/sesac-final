```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1. 브라우저 설정
url = "https://poomgo.com/blog/e-commerce/kbeauty-japan-crossborder-trends-2025"
options = Options()
# options.add_argument("--headless") # 창 없이 실행하려면 주석 해제
driver = webdriver.Chrome(options=options)

try:
    print(f"🔄 페이지 접속 및 데이터 렌더링 대기 중...")
    driver.get(url)

    # 2. 동적 컨텐츠 로딩 대기 (최대 15초)
    wait = WebDriverWait(driver, 15)
    # 클래스명에 'PostLayout_content_section'이 포함된 div가 나타날 때까지 기다림
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div[class*="PostLayout_content_section"]')))

    # 전체 페이지 파싱
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # 3. [핵심] 부분 일치 선택자 사용
    # 사진에서 본 'PostLayout_content_section___7TpL' 중 앞부분인 'PostLayout_content_section'으로 탐색
    target_selector = 'div[class*="PostLayout_content_section"][class*="StrapiContent_strapi_content"]'
    content_area = soup.select_one(target_selector)

    if content_area:
        # 4. 전문 추출 (가독성을 위해 줄바꿈 삽입)
        full_text = content_area.get_text(separator='\n', strip=True)

        print("\n" + "🚀" * 20)
        print("✅ 데이터 수집 성공! 전문을 출력합니다.")
        print("🚀" * 20 + "\n")
        
        print(full_text)
        
        print("\n" + "=" * 40)
        print("💡 위 내용을 복사해서 프롬프트에 활용하세요.")
        print("=" * 40)

    else:
        # 실패 시 디버깅을 위해 현재 페이지의 모든 클래스명 출력 (선택 사항)
        print("❌ 여전히 클래스 구역을 찾지 못했습니다. 구조가 변경되었을 수 있습니다.")
        
except Exception as e:
    print(f"❌ 오류 발생: {e}")

finally:
    driver.quit()
```
