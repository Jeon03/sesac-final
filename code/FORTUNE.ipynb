```python
# 필요 라이브러리 임포트
import requests
from bs4 import BeautifulSoup

# 1. 타겟 URL 및 헤더 설정 (봇 차단 방지)
url = "https://www.fortunebusinessinsights.com/ko/japan-cosmetics-market-114215"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# 2. 웹페이지에 GET 요청 전송
response = requests.get(url, headers=headers)

# 3. 접속 성공(상태 코드 200) 확인 및 데이터 파싱
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 4. CSS 선택자를 활용하여 다중 클래스를 가진 div 요소 찾기
    target_selector = 'div.col-lg-12.healthcare.no-border.ltr-class'
    target_div = soup.select_one(target_selector)
    
    # 5. 특정 태그만 위에서부터 순서대로 추출
    if target_div:
        # 찾고자 하는 태그들을 리스트로 정의합니다.
        target_tags = ['h1', 'h2', 'h3', 'p', 'li', 'a']
        
        # find_all은 HTML 문서의 위에서 아래 순서(Top-down)로 요소를 찾습니다.
        elements = target_div.find_all(target_tags)
        
        extracted_lines = []
        for elem in elements:
            # 각 태그에서 텍스트를 추출하고 좌우 공백을 제거합니다.
            text = elem.get_text(strip=True)
            # 빈 텍스트가 아닌 경우에만 리스트에 추가합니다.
            if text: 
                extracted_lines.append(text)
        
        # 추출된 텍스트 리스트를 줄바꿈(\n)으로 연결하여 최종 문자열 생성
        final_text = '\n'.join(extracted_lines)
        
        print("✅ 특정 태그 데이터 수집 성공!\n")
        print("-" * 50)
        print(final_text) 
        print("-" * 50)
        
    else:
        print("❌ 해당 클래스를 가진 div 요소를 페이지에서 찾을 수 없습니다.")
else:
    print(f"❌ 웹페이지 접속 실패: 상태 코드 {response.status_code}")
```
