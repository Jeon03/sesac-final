```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
import random

# 1. 드라이버 체크 및 실행
if 'driver' not in globals() or driver is None:
    print("🌐 드라이버를 새로 시작합니다...")
    options = uc.ChromeOptions()
    driver = uc.Chrome(version_main=145, options=options)
    print("⚠️ 로그인이 필요한 경우 완료 후 엔터를 눌러주세요.")
    input() 
else:
    print("✅ 기존 브라우저 세션을 사용합니다.")

# 2. 번역된 상품 정보 및 URL 설정
# 원문: Bánh Ống Gạo Lứt Bebecook Bánh Gạo Ăn Dặm Thơm Ngon Nhiều Vị Dễ Tan
translated_product_name = "베베쿡 현미 롱스틱 쌀과자 (다양한 맛, 잘 녹는 제형)"
target_url = "https://shopee.vn/B%C3%A1nh-%E1%BB%90ng-G%E1%BA%A1o-L%E1%BB%A9t-Bebecook-B%C3%A1nh-G%E1%BA%A1o-%C4%82n-D%E1%BA%B7m-Th%C6%A1m-Ngon-Nhi%E1%BB%81u-V%E1%BB%8B-D%E1%BB%85-Tan-i.595735611.16585075040"
target_count = 22
third_reviews = []

# 페이지 이동
driver.get(target_url)
print(f"🚀 '{translated_product_name}' 수집을 시작합니다...")
time.sleep(5) 

# 

# 3. 데이터 수집 루프
try:
    while len(third_reviews) < target_count:
        # 안전 장치: 리뷰 섹션 로딩을 위한 단계적 스크롤
        for _ in range(5):
            driver.execute_script("window.scrollBy(0, 600);")
            time.sleep(0.8)
        
        try:
            WebDriverWait(driver, 15).until(
                EC.presence_of_element_located((By.CLASS_NAME, "q2b7Oq"))
            )
        except:
            print("\n⚠️ 리뷰 로딩 지연. 수동으로 스크롤을 조절해 보세요.")
            time.sleep(3)
            continue

        items = driver.find_elements(By.CLASS_NAME, "q2b7Oq")
        
        for item in items:
            try:
                author = item.find_element(By.CLASS_NAME, "InK5kS").text
                stars = len(item.find_elements(By.CLASS_NAME, "icon-rating-solid"))
                date_info = item.find_element(By.CLASS_NAME, "XYk98l").text
                content = item.find_element(By.CLASS_NAME, "YNedDV").text
                
                date_val = date_info.split('|')[0].strip() if '|' in date_info else date_info
                variation_val = date_info.split('|')[1].replace('Variation:', '').strip() if '|' in date_info else 'N/A'

                third_reviews.append({
                    '상품명': translated_product_name,
                    '작성자': author,
                    '별점': stars,
                    '날짜': date_val,
                    '옵션': variation_val,
                    '내용': content.replace('\n', ' ')
                })
                if len(third_reviews) >= target_count: break
            except:
                continue
        
        print(f"📊 수집 진행: {len(third_reviews)} / {target_count}", end='\r')
        if len(third_reviews) >= target_count: break

        # 다음 버튼 클릭 (SVG 화살표)
        try:
            next_btn = driver.find_element(By.CLASS_NAME, "shopee-icon-button--right")
            if next_btn.get_attribute("disabled") is not None: break
            driver.execute_script("arguments[0].click();", next_btn)
            time.sleep(random.uniform(3.0, 5.0)) 
        except:
            break

    # 4. JSON 저장
    with open('베베쿡 현미 롱스틱 쌀과자 (다양한 맛, 잘 녹는 제형).json', 'w', encoding='utf-8') as f:
        json.dump(third_reviews, f, ensure_ascii=False, indent=4)
    
    print(f"\n\n🎉 수집 성공! '{translated_product_name}' 데이터 저장 완료.")

except Exception as e:
    print(f"\n🚨 에러 발생: {e}")
```

```python
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import json
import random

# 1. [드라이버 재시작 및 체크 로직]
def get_driver():
    global driver
    try:
        # 기존 드라이버가 있고 창이 열려있는지 확인
        _ = driver.window_handles
        print("✅ 기존 브라우저 세션을 그대로 사용합니다.")
    except (NameError, Exception):
        print("🌐 드라이버가 없거나 닫혀 있습니다. 새로 시작합니다...")
        options = uc.ChromeOptions()
        # 여욱님의 크롬 버전 145에 맞춰 실행
        driver = uc.Chrome(version_main=145, options=options)
        print("⚠️ 브라우저가 열렸습니다. 라자다 로그인을 완료한 후 엔터를 눌러주세요.")
        input() 
    return driver

# 드라이버 확보
driver = get_driver()

# 2. 상품 정보 및 설정
translated_product_name = "General Summary of Korean Bebecook Dietary Cakes for Baby"
target_url = "https://www.lazada.vn/products/pdp-i1608412199-s8082460909.html" # 라자다 상품 주소
target_count = 41 # 총 리뷰 수
all_reviews = []

# 페이지 이동
driver.get(target_url)
print(f"🚀 '{translated_product_name}' 수집을 시작합니다...")
time.sleep(5)

# 

# 3. 데이터 수집 루프
try:
    while len(all_reviews) < target_count:
        # 리뷰 섹션이 보일 때까지 스크롤 및 대기
        WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.CLASS_NAME, "mod-reviews"))
        )
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight * 0.5);")
        time.sleep(2)

        items = driver.find_elements(By.CLASS_NAME, "item")
        
        for item in items:
            try:
                author = item.find_element(By.CLASS_NAME, "reviewer").text
                review_time = item.find_element(By.CLASS_NAME, "time").text
                
                # 별점 추출 (색깔이 채워진 별 개수 확인)
                stars = item.find_elements(By.CSS_SELECTOR, "path[style*='fill: rgb(255, 200, 60)']")
                rating = len(stars) // 2 if len(stars) > 0 else 5

                # 상품 옵션 및 내용
                try:
                    option = item.find_element(By.CLASS_NAME, "skuInfo-value").text
                except:
                    option = "N/A"
                
                content = item.find_element(By.CLASS_NAME, "item-content-main-content-reviews").text
                
                all_reviews.append({
                    'platform': 'Lazada',
                    'product_name': translated_product_name,
                    'author': author,
                    'rating': rating,
                    'date': review_time,
                    'variation': option,
                    'content': content.replace('\n', ' ')
                })
                if len(all_reviews) >= target_count: break
            except:
                continue
        
        print(f"📊 수집 진행: {len(all_reviews)} / {target_count}", end='\r')
        if len(all_reviews) >= target_count: break

        # 다음 페이지 클릭 (iweb-pagination-next)
        try:
            next_btn = driver.find_element(By.CLASS_NAME, "iweb-pagination-next")
            if "iweb-pagination-disabled" in next_btn.get_attribute("class"):
                print("\n🏁 마지막 페이지에 도달했습니다.")
                break
                
            driver.execute_script("arguments[0].click();", next_btn)
            time.sleep(random.uniform(3.0, 5.0)) 
        except:
            print("\n⚠️ 다음 페이지 버튼을 찾을 수 없습니다.")
            break

    # 4. JSON 저장
    output_file = 'bebecook_lazada_final.json'
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(all_reviews, f, ensure_ascii=False, indent=4)
    print(f"\n\n🎉 수집 성공! '{output_file}'에 저장 완료.")

except Exception as e:
    print(f"\n🚨 에러 발생: {e}")
```
